```python
import numpy as np
import scipy.io as sio
data_dir = '/data2/share/multimodal_dataset/MS COCO/'
pro_dir = '/data2/knif/dataset/mscoco/'
ind = sio.loadmat(data_dir + 'index.mat')

indR = ind['indR']
indT = ind['indT']
indQ = ind['indQ']

def save():
    img = np.load(data_dir + 'IAll.npy')
    txt = sio.loadmat(data_dir + 'ms-coco-yall.mat')['YAll']
    label = sio.loadmat(data_dir + 'ms-coco-lall.mat')['LAll']
    feat = sio.loadmat(data_dir + 'mscoco_vgg19.mat')['mscoco_vgg19']

    # original
    np.savez(pro_dir + 'original.npz', img=img, txt=txt, label=label, feat=feat)
    # dataset
    np.savez(pro_dir + 'database.npz', img=img[indR], txt=txt[indR], label=label[indR], feat=feat[indR])
    # train
    np.savez(pro_dir + 'train.npz', img=img[indT], txt=txt[indT], label=label[indT], feat=feat[indT])
    # test
    np.savez(pro_dir + 'test.npz', img=img[indQ], txt=txt[indQ], label=label[indQ], feat=feat[indQ])
    # ind
    np.savez(pro_dir + 'ind.npz', database=indR, train=indT, test=indQ)

save()
```
